# Colors...

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_all_matplotlib_cmaps():
    # Get the list of all available colormaps
    cmaps = plt.colormaps()

    # Create some data to display the colormaps
    gradient = np.linspace(0, 1, 256)
    gradient = np.vstack((gradient, gradient))

    # Define how many colormaps to show per row
    n_col = 4
    n_row = len(cmaps) // n_col + 1

    # Create a figure to display the colormaps
    fig, axes = plt.subplots(n_row, n_col, figsize=(12, 15))
    axes = axes.flatten()

    # Loop through all colormaps and display them
    for i, cmap in enumerate(cmaps):
        ax = axes[i]
        ax.imshow(gradient, aspect='auto', cmap=plt.get_cmap(cmap))
        ax.set_title(cmap, fontsize=8)
        ax.axis('off')

    # Turn off any unused subplots
    for j in range(i+1, len(axes)):
        axes[j].axis('off')

    plt.tight_layout()
    plt.show()

# Call the function
plot_all_matplotlib_cmaps()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

def plot_all_seaborn_palettes():
    # List of common Seaborn palettes
    palettes = [
        "deep", "muted", "bright", "pastel", "dark", "colorblind",
        "husl", "hsv", "Paired", "Set1", "Set2", "Set3", "tab10", "tab20"
    ]
    
    # Create a figure to display the color palettes
    n_col = 2  # Palettes per row
    n_row = len(palettes) // n_col + 1
    fig, axes = plt.subplots(n_row, n_col, figsize=(10, 8))
    axes = axes.flatten()

    # Loop through the palettes and display them
    for i, palette in enumerate(palettes):
        ax = axes[i]
        colors = sns.color_palette(palette, 10)  # Get the palette with 10 colors
        # Plot the colors as a series of rectangles (like palplot)
        for j, color in enumerate(colors):
            ax.add_patch(plt.Rectangle((j, 0), 1, 1, color=color))
        
        ax.set_xlim(0, len(colors))
        ax.set_ylim(0, 1)
        ax.set_title(palette, fontsize=12)
        ax.axis('off')

    # Turn off any unused subplots
    for j in range(i + 1, len(axes)):
        axes[j].axis('off')

    plt.tight_layout()
    plt.show()

# Call the function
plot_all_seaborn_palettes()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.cm as cm

# Sample data
data = pd.DataFrame({
    'category': ['A', 'B', 'C', 'A', 'B', 'C'],
    'value': [5, 7, 3, 6, 5, 9],
    'group': ['G1', 'G1', 'G1', 'G2', 'G2', 'G2']
})

# Create a colormap with Matplotlib
cmap = cm.get_cmap('viridis', 2)  # Using 2 colors for the hue (for 2 groups)

# Convert the colormap into a Seaborn palette
sns_palette = sns.color_palette([cmap(i) for i in range(cmap.N)])

# Create a Seaborn barplot with the custom colormap palette
sns.barplot(x='category', y='value', hue='group', data=data, palette=sns_palette)

# Display the plot
plt.title('Seaborn Barplot with Matplotlib Colormap for Legend')
plt.show()
```

# Error bars...

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 10)
y = np.sin(x)
yerr = 0.2

plt.errorbar(x, y, yerr=yerr, fmt='o', ecolor='red', elinewidth=2, capsize=5, capthick=2, alpha=0.7)
plt.show()
```

Key Aesthetic Parameters in plt.errorbar():

	•	xerr and yerr: Specifies the error for x and y data points.
	•	fmt: The format string for the plot (e.g., marker style, line style).
	•	ecolor: The color of the error bars (accepts any valid color format in Matplotlib).
	•	elinewidth: The width of the error bar lines.
	•	capsize: The length of the caps on the error bars (set to 0 for no caps).
	•	capthick: The thickness of the caps on the error bars.
	•	errorevery: Set how often to draw the error bars (e.g., errorevery=2 draws every second error bar).
	•	alpha: The transparency of the error bars (between 0 and 1).
	•	zorder: Controls the drawing order of the error bars (higher values appear in front).
	•	barsabove: If True, error bars are drawn above the plot elements, otherwise below.

```python
# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

sns.lineplot(x=x, y=y, errorbar='sd', err_style="bars", err_kws={"capsize": 5, "elinewidth": 2})
plt.show()
```

Key Aesthetic Parameters in Seaborn:

	•	ci: Confidence interval to plot. If None, no error bars are drawn. Default is 95 (for 95% confidence interval). Set ci="sd" to display the standard deviation.
	•	errcolor: The color of the error bars.
	•	errwidth: The thickness of the error bars.
	•	capsize: The width of the caps on the error bars.
	•	alpha: Transparency level for error bars.
	•	n_boot: Number of bootstrap resamples to compute confidence intervals.
	•	errorbar (for new Seaborn 0.12+): Defines the type of error bars to draw, e.g., "sd", "ci", "pi", "se", or "sd"
